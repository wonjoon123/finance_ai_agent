```python
import pandas as pd
import numpy as np

def calculate_rsi_vectorized(df_close, period=14):
    delta = df_close.diff(axis=1)
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    gain_ma = pd.DataFrame(gain, index=df_close.index, columns=df_close.columns).rolling(window=period, axis=1).mean()
    loss_ma = pd.DataFrame(loss, index=df_close.index, columns=df_close.columns).rolling(window=period, axis=1).mean()
    rs = gain_ma / loss_ma
    rsi = 100 - (100 / (1 + rs))
    return rsi

target_date = "{ÎÇ†Ïßú}"
threshold = "{ÏàòÏπò}"

date_cols = [c for c in stock_data.columns if c not in ['Price','Ticker','ÏãúÏû•','Ï¢ÖÎ™©Î™Ö']]
close_data = stock_data[stock_data['Price'] == 'Close'].set_index('Ticker')
close_only = close_data[date_cols].astype(float)

rsi_df = calculate_rsi_vectorized(close_only)
filtered = rsi_df[rsi_df[target_date] >= threshold]

result = [
    f"{close_data.loc[ticker, 'Ï¢ÖÎ™©Î™Ö']} (RSI:{rsi_df.loc[ticker, target_date]:.1f})"
    for ticker in filtered.index
]

result
```

```python
rolling_window = 20
target_date = '2025-02-17'

# ÎÇ†Ïßú Ïª¨ÎüºÎßå Ï∂îÏ∂ú (Ïïû 2Í∞ú, Îí§ 2Í∞ú Ï†úÏô∏)
date_cols = stock_data.columns[2:-2]

# Í±∞ÎûòÎüâ Îç∞Ïù¥ÌÑ∞ ÌïÑÌÑ∞ÎßÅ
volume_data = stock_data[stock_data['Price'] == 'Volume'].copy()

# 20Ïùº Ïù¥ÎèôÌèâÍ∑† Í≥ÑÏÇ∞
volume_only = volume_data[date_cols].T
ma20 = volume_only.rolling(window=rolling_window, min_periods=rolling_window).mean().T

# Ìï¥Îãπ ÎÇ†ÏßúÏùò Í±∞ÎûòÎüâÍ≥º 20Ïùº ÌèâÍ∑† Í±∞ÎûòÎüâ Ï†ÄÏû•
volume_data['MA20'] = ma20[target_date]
volume_data['Í±∞ÎûòÎüâ'] = volume_data[target_date]

# Í∏âÏ¶ùÎ•† Í≥ÑÏÇ∞: (Í±∞ÎûòÎüâ / MA20) * 100
volume_data['Í∏âÏ¶ùÎ•†'] = (volume_data['Í±∞ÎûòÎüâ'] / volume_data['MA20']) * 100

# 300% Ïù¥ÏÉÅ Í∏âÎì±Ìïú Ï¢ÖÎ™© ÌïÑÌÑ∞ÎßÅ
filtered = volume_data[volume_data['Í∏âÏ¶ùÎ•†'] >= 300].copy()

# Í∏âÏ¶ùÎ•† ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨
filtered = filtered.sort_values(by='Í∏âÏ¶ùÎ•†', ascending=False)

# Ìè¨Îß∑: Ï¢ÖÎ™©Î™Ö(1234%)
result = filtered.apply(
    lambda row: f"{row['Ï¢ÖÎ™©Î™Ö']}({int(round(row['Í∏âÏ¶ùÎ•†']))}%)", axis=1
)

# Ï∂úÎ†•
print(", ".join(result))
```
